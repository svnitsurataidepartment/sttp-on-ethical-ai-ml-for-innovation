```python
from datasets import load_dataset

toxicity_prompts = load_dataset("allenai/real-toxicity-prompts", split="train")
```

```python
toxicity_prompts[43]['prompt']
```

```python
# randomly select some prompts
from random import sample
all_indexes = list(range(len(toxicity_prompts)))
sample_size = 50
sample_indexes = sample(all_indexes, sample_size)
full_prompt_dict = toxicity_prompts[sample_indexes]
only_prompts = [prompt['text'] for prompt in full_prompt_dict['prompt']]
only_prompts
```

```python
from transformers import pipeline, AutoTokenizer

text_generation = pipeline("text-generation", model="gpt2")
tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
model_continuations = []
for prompt in only_prompts:
  generation = text_generation(prompt, max_length=100, do_sample=False, pad_token_id=50256)
  continuation = generation[0]['generated_text'].replace(prompt,'')
  model_continuations.append('Initial Prompt: ' + prompt + '\n' + 'generated continuation:' + continuation)

print('Generated '+ str(len(model_continuations))+ ' continuations')
```

```python
model_continuations
```

```python
import evaluate
toxicity = evaluate.load("toxicity")
```

```python
toxicity_ratio = toxicity.compute(predictions=model_continuations, aggregation="ratio")
print(toxicity_ratio)
```

```python
max_toxicity = toxicity.compute(predictions=model_continuations, aggregation="maximum")
print(max_toxicity)
```

```python
# text wise toxicity
tox_dict = {}
all_toxicity = toxicity.compute(predictions=model_continuations)
for text, score in zip(model_continuations, all_toxicity['toxicity']):
  tox_dict[text] = score
```

```python
tox_dict = (dict(sorted(tox_dict.items(), key=lambda item: item[1], reverse=True)))
```
